```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([0,4])
y=np.array([0,4])

plt.plot(x,y)
plt.show()
```

```python
x = np.array([0,1,2,3,4])
y = np.array([0,1,2,3,4])
             
plt.plot(x,y, marker = "*") 
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
x = np.array(["apples","bananas","mango","grapes"])
y = np.array([4,15,20,2])

plt.bar(x,y,color="yellow",width =0.4)
plt.show()
```
